## 折れ線グラフ
[matplotlib.pyplot.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html?highlight=plot#matplotlib.pyplot.plot)

```python
# データの用意
x = [1, 2, 3]
y1 = [1, 2, 3]
y2 = [3, 1, 2]

# 折れ線グラフの描画
fig, ax = plt.subplots()
ax.plot(x, y1)
ax.plot(x, y2)
plt.show()
```

```python
# データの用意
x = np.arange(0.0, 15.0, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# 折れ線グラフの描画
fig, ax = plt.subplots()
ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')
ax.legend()
plt.show()
```

## 棒グラフ
[matplotlib.pyplot.bar](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html)

```python
# データの用意
x = [1, 2, 3]
y = [10, 2, 3]

# 棒グラフの描画
fig, ax = plt.subplots()
labels = ['spam', 'ham', 'egg']
ax.bar(x, y, tick_label=labels)
plt.show()
```

### 横向きの棒グラフ
[matplotlib.pyplot.barh](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.barh.html?highlight=barh#matplotlib.pyplot.barh)

```python
# データの用意
x = [1, 2, 3]
y = [10, 2, 3]

# 横向き棒グラフの描画
fig, ax = plt.subplots()
labels = ['spam', 'ham', 'egg']
ax.barh(x, y, tick_label=labels)
plt.show()
```

### 2つの棒グラフを並べる

```python
# データの用意
x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]
labels = ['spam', 'ham', 'egg']

# y1のグラフを描画(幅を0.4にする)
fig, ax = plt.subplots()
width = 0.4
ax.bar(x, y1, width=width, tick_label=labels, label='y1')

# y2のグラフを描画(幅分だけxをズラす)
x2 = [num + width for num in x]
ax.bar(x2, y2, width=width, label='y2')

ax.legend()
plt.show()
```

### 積み上げ棒グラフ

```python
# データの用意
x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]
labels = ['spam', 'ham', 'egg']

# y1とy2を合わせた値を用意
y_total = [num1 + num2 for num1, num2 in zip(y1, y2)]

# 積み上げ棒グラフの描画
fig, ax = plt.subplots()
ax.bar(x, y_total, tick_label=labels, label='y1')
ax.bar(x, y2, label='y2')
ax.legend()
plt.show()
```

## 散布図
[matplotlib.pyplot.scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)

```python
# データの用意
np.random.seed(123)
x = np.random.rand(20)
y = np.random.rand(20)

# 散布図の描画
fig, ax = plt.subplots()
ax.scatter(x[:10], y[:10], marker='x', label='class 0')
ax.scatter(x[10:], y[10:], marker='o', label='class 1')
ax.legend()
plt.show()
```

## ヒストグラフ
[matplotlib.pyplot.hist](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html?highlight=hist#matplotlib.pyplot.hist)

```python
# データの用意
np.random.seed(123)
mu, sigma = 100, 15
x = np.random.normal(mu, sigma, 1000)

# binsパラメーターで細かさを指定
fig, ax = plt.subplots()
n, bins, patches = ax.hist(x, bins=25)    # デフォルトのbin数は「10」 25個の要素を作成するため、26個の値(度数)を作成される
plt.show()
```

```python
# bins内には各binの境界の値(つまりX軸のデータ)、nには各binの度数(要素数、つまりY軸のデータ)が入っているので、以下のように数値を取得できる。ちなみに、patchesは、ヒストグラム各棒のオブジェクトのリストになっている
for i, num in enumerate(n):
    print('{:.2f} - {:.2f}: {}'.format(bins[i], bins[i+1], num))
```

### 横向きのヒストグラフ

```python
# データの用意
np.random.seed(123)
mu, sigma = 100, 15
x = np.random.normal(mu, sigma, 1000)

# orientation=horizontalと指定することで横向き
fig, ax = plt.subplots()
n, bins, patches = ax.hist(x, orientation='horizontal')
plt.show()
```

### 複数の棒グラフを利用したヒストグラフ

```python
# データの用意
np.random.seed(123)
x = np.random.normal(loc=100, scale=20, size=(1000))
y = np.random.normal(loc=100, scale=15, size=(1000))
z = np.random.normal(loc=100, scale=10, size=(1000))

# 複数の棒グラフを利用したヒストグラフ
fig, ax = plt.subplots()
labels = ['x', 'y', 'z']
ax.hist((x, y, z), label=labels)
ax.legend()
plt.show()
```

### 積み上げヒストグラフ

```python
# データの用意
np.random.seed(123)
x = np.random.normal(loc=100, scale=20, size=(1000))
y = np.random.normal(loc=100, scale=15, size=(1000))
z = np.random.normal(loc=100, scale=10, size=(1000))

# 積み上げヒストグラフ
fig, ax = plt.subplots()
labels=['x', 'y', 'z']
ax.hist((x, y, z), label=labels, stacked=True)
ax.legend()
plt.show()
```

## 箱ひげ図
[matplotlib.pyplot.boxplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html?highlight=boxplot#matplotlib.pyplot.boxplot)

|||
|---|---|
| 上ひげ | 最大値 |
| 箱の上部 | 第３四分位数 |
| 箱内の赤線 | 中央値 |
| 箱の下部 | 第１四分位数 |
| 下ひげ | 最小値 |
| まる | 外れ値 |

```python
# データの用意
np.random.seed(123)
x = np.random.normal(loc=0, scale=10, size=(500))
y = np.random.normal(loc=0, scale=15, size=(500))
z = np.random.normal(loc=0, scale=20, size=(500))

# 箱ひげ図の描画
fig, ax = plt.subplots()
labels = ['x', 'y', 'z']
ax.boxplot((x, y, z), labels=labels)
plt.show()
```

```python
# データの用意
np.random.seed(123)
x = np.random.normal(loc=0, scale=10, size=(500))
y = np.random.normal(loc=0, scale=15, size=(500))
z = np.random.normal(loc=0, scale=20, size=(500))

# vert=Falseで横向き
fig, ax = plt.subplots()
labels = ['x', 'y', 'z']
ax.boxplot((x, y, z), labels=labels, vert=False)
plt.show()
```

## 円グラフ
[matplotlib.pyplot.pie](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html?highlight=pie#matplotlib.pyplot.pie)

```python
# データの用意
x = [10, 3, 1]
labels = ['spam', 'ham', 'egg']

# 円グラフの描画
fig, ax = plt.subplots()
ax.pie(x, labels=labels, startangle=90, counterclock=False)  # 上(90度)から時計回りを指定
ax.axis('equal')                                             # アスペクト比を保持
plt.show()
```

```python
# データの用意
x = [10, 3, 1]
labels = ['spam', 'ham', 'egg']

# 円グラフの描画
fig, ax = plt.subplots()
ax.pie(x, labels=labels, startangle=90, counterclock=False, shadow=True, autopct='%1.2f%%', pctdistance=0.8)    # 影付けして、パーセントを表記、pctdistanceでパーセント表記の場所を定義(1以上で円外に記載)
ax.axis('equal')
plt.show()
```

```python
# データの用意
x = [10, 3, 1]
labels = ['spam', 'ham', 'egg']
explode = [0, 0.2, 0]

# 円グラフの描画
fig, ax = plt.subplots()
ax.pie(x, labels=labels, startangle=90, counterclock=False, explode=explode)    # explodeを指定
ax.axis('equal')
plt.show()
```

## フォントのスタイル

```python
fig, ax = plt.subplots()

ax.set_xlabel('xlabel', family='sans-serif', size=20, weight='bold')
ax.set_ylabel('ylabel', family='cursive', size=40, weight='light')
ax.set_title('graph title', family='monospace', size=25, weight='heavy')

plt.show()
```

```python
# フォントのスタイルを辞書で定義
fontdict = {
    'family': 'monospace',
    'size': 20,
    'weight': 'normal'
}

fig, ax = plt.subplots()

ax.set_xlabel('xlabel', fontdict=fontdict)
ax.set_ylabel('ylabel', fontdict=fontdict)
# 上書き
ax.set_title('graph title', fontdict=fontdict, size=40)

plt.show()
```
